```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#设置seaborn参数
sns.set_style('whitegrid')
sns.set_style({'font.sans-serif':['simhei', 'Arial']})
sns.set_palette(sns.color_palette('BuGn_d'))
```

```python
file_name = r'./电机堵转数据.xlsx'
#读入数据
df_all = pd.read_excel(file_name, sheet_name=[0, 1])
#对数据按驱动电机转速进行升序排序
df = df_all[0]
df
```

```python
df_2 = df_all[1]
df_2
```

```python
Pk = list(map(lambda x,y : x*y, df['相电压'],df['相电流']))
Pk
```

```python
Po = list(map(lambda x,y : x*y, df_2['相电压'],df_2['相电流']))
Po 
```

```python
plt.plot(df['线电压'],df['相电流'], label='Uo - Ik', color= 'r')
# plt.plot(df['电压'],Pk, label = 'Uo - Pk', color='c')
plt.legend()
plt.title('短路（堵转）电流曲线')
plt.xlabel("Uo/V")
plt.ylabel("Ik/A")
```

```python
plt.plot(df['线电压'],Pk, label = 'Uo - Pk', color='c')
plt.legend()
plt.title('短路（堵转）功率曲线')
plt.xlabel("Uo/V")
plt.ylabel("Pk/W")
```

```python
plt.plot(df_2['线电压'],df_2['相电流'], label='Uo - Io', color= 'r')
# plt.plot(df['电压'],Pk, label = 'Uo - Pk', color='c')
plt.legend()
plt.title('空载电流曲线')
plt.xlabel("Uo/V")
plt.ylabel("Io/A")
```

```python
plt.plot(df_2['线电压'],Po, label = 'Uo - Po', color='c')
plt.legend()
plt.title('空载功率曲线')
plt.xlabel("Uo/V")
plt.ylabel("Po/W")
```
