```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#设置seaborn参数
sns.set_style('whitegrid')
sns.set_style({'font.sans-serif':['simhei', 'Arial']})
sns.set_palette(sns.color_palette('BuGn_d'))
```

```python
file_name = r'./机电实验三数据.xlsx'
#读入数据
df_all = pd.read_excel(file_name, sheet_name=[0, 1, 2])
```

```python
df_all[0]
```

```python
plt.plot(df_all[0]['输入电压(V)'],df_all[0]['输入电流(A)'], label='输入电压 - 电流', color= 'r')
plt.legend()
plt.title('空载特性-电流曲线')
plt.xlabel("输入电压")
plt.ylabel("输入电流")
```

```python
plt.plot(df_all[0]['输入电压(V)'],df_all[0]['输入功率(KW)'], label='输入电压 - 功率', color= 'g')
plt.legend()
plt.title('空载特性-功率曲线')
plt.xlabel("输入电压")
plt.ylabel("功率")
```

```python
plt.plot(df_all[0]['输入电压(V)'],df_all[0]['输出转速(rpm)'], label='输入电压 - 输出转速', color= 'c')
plt.legend()
plt.title('空载特性-转速曲线')
plt.xlabel("输入电压")
plt.ylabel("转速")
```
