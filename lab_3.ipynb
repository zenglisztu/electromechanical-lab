```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#设置seaborn参数
sns.set_style('whitegrid')
sns.set_style({'font.sans-serif':['simhei', 'Arial']})
sns.set_palette(sns.color_palette('BuGn_d'))
```

```python
file_name = r'./机电实验三数据.xlsx'
#读入数据
df_all = pd.read_excel(file_name, sheet_name=[0, 1, 2])
```

```python
df_all[0]
```

```python
fig = plt.figure(figsize=(8, 6))
plt.plot(df_all[0]['输入电压(V)'],df_all[0]['输入电流(A)'], label='输入电压 - 电流', color= 'r', marker='o',markersize=8, markeredgecolor='yellow')
for i, (xx, yy) in enumerate(zip(df_all[0]['输入电压(V)'], df_all[0]['输入电流(A)'])):
    plt.text(xx, yy, f'({xx:.2f}, {yy:.2f})', ha='center', va='bottom')
plt.legend()
plt.title('空载特性 电压-电流曲线')
plt.xlabel("输入电压(V)")
plt.ylabel("输入电流(A)")
```

```python
plt.plot(df_all[0]['输入电压(V)'],df_all[0]['输出功率(kw)'], label='输入电压 - 功率', color= 'y', marker='o',markersize=8, markeredgecolor='yellow')
for i, (xx, yy) in enumerate(zip(df_all[0]['输入电压(V)'], df_all[0]['输出功率(kw)'])):
    plt.text(xx, yy, f'({xx:.2f}, {yy:.2f})', ha='center', va='bottom')
plt.legend()
plt.title('空载特性 电压-功率曲线')
plt.xlabel("输入电压(V)")
plt.ylabel("输出功率(KW)")
```

```python
plt.plot(df_all[0]['输入电压(V)'],df_all[0]['输出转速(rpm)'], label='输入电压 - 输出转速', color= 'c', marker='o',markersize=8, markeredgecolor='yellow')
for i, (xx, yy) in enumerate(zip(df_all[0]['输入电压(V)'], df_all[0]['输出转速(rpm)'])):
    plt.text(xx, yy, f'({xx:.2f}, {yy:.2f})', ha='center', va='bottom')
plt.legend()
plt.title('空载特性 电压-转速曲线')
plt.xlabel("输入电压(V)")
plt.ylabel("输出转速(rpm)")
```
