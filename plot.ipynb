```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
#设置seaborn参数
sns.set_style('whitegrid')
sns.set_style({'font.sans-serif':['simhei', 'Arial']})
sns.set_palette(sns.color_palette('BuGn_d'))
```

```python
file_name = r'./机电实验数据.xlsm'
#读入数据
df_all = pd.read_excel(file_name, sheet_name=[0, 1, 2])
#对数据按驱动电机转速进行升序排序
df = df_all[0]
df = df.sort_values('驱动电机转速（r/min）')
df
```

```python
#负载力矩为额定力矩
sns.relplot(x='输入频率(Hz)', y='输入电压(V)', kind='line', data=df)
plt.title('额定力矩 电压与频率曲线')
```

```python
#对数据按驱动电机转速进行升序排序
df_half = df_all[1]
df_half = df_half.sort_values('驱动电机转速（r/min）')
df_half
```

```python
#负载力矩为0.5倍力矩
sns.relplot(x='输入频率(Hz)', y='输入电压(V)', kind='line', data=df_half)
plt.title('0.5倍力矩力矩 电压与频率曲线')
```

```python
#改变负载力矩设定，维持转速设定为额定转速
df_f = df_all[2]
#对数据按负载电机转矩进行升序排序
df_f = df_f.sort_values('负载电机转矩(N.m)')
df_f
```

```python
sns.relplot(y='负载电机转矩(N.m)', x='输入电流(A)', kind='line', data=df_f)
plt.title('力矩与电流曲线')
```

```python
#比较上述两组实验（1）（2）
plt.plot(df['输入频率(Hz)'],df['输入电压(V)'], label='额定力矩', color= 'r')
plt.plot(df_half['输入频率(Hz)'],df_half['输入电压(V)'], label = '0.5倍额定力矩', color='c')
plt.legend()
plt.title('电压与频率曲线')
plt.xlabel("输入频率(Hz)")
plt.ylabel("输入电压(V)")
```
